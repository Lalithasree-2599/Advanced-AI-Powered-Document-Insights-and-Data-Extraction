```python
# ==============================
# JSON ‚Üí Flat DataFrame ‚Üí JSON
# ==============================

import json
import pandas as pd

# ---- 1Ô∏è‚É£ Load the original JSON ----
with open('/content/sample.json', 'r') as file:
    data = json.load(file)

# ---- 2Ô∏è‚É£ Normalize the JSON structure ----
records = []
for customer in data:
    for order in customer['orders']:
        for item in order['items']:
            records.append({
                'customer_id': customer['customer_id'],
                'order_id': order['order_id'],
                'product': item['product'],
                'quantity': item['quantity'],
                'total': order['total']
            })

# ---- 3Ô∏è‚É£ Create DataFrame ----
df = pd.DataFrame(records)

# ---- 4Ô∏è‚É£ Save as a new JSON file ----
output_file = 'flattened_orders.json'
df.to_json(output_file, orient='records', indent=4)

# ---- 5Ô∏è‚É£ Display confirmation ----
print(f"‚úÖ Flattened data saved to: {output_file}")

# ---- 6Ô∏è‚É£ Generate download link (Colab only) ----
from google.colab import files
files.download(output_file)
```

```python
# Standardizing common terms
common_terms = {
    "Dr.": "Doctor",
    "NYC": "New York",
    "N.Y.": "New York",
    "USD": "dollars",
    "$": "dollars"
}

def standardize_terms(text, replacements):
    for key, value in replacements.items():
        text = text.replace(key, value)
    return text

# Example
text = "Dr. Smith is based in NYC and charges 100 USD."
clean_text = standardize_terms(text, common_terms)
print(clean_text)
# Output: "Doctor Smith is based in New York and charges 100 dollars."
```

```python
import re
import string

text = """Th√≠s √≠s √† pr√≥blem√°tic t√©xt f√≠le!! It contains    **extra spaces** ,,,,, special characters!!!üí•üî•üöÄ
Some words are misspelleddd,   and encoding  issues   lik√© th√≠s c√§us√© probl√´ms.
Prices are inconsistent:  $29.99, 29.99 USD, 29,99$.
Emails & phone numbers may be embedded: contact@domain.com, (123)-456-7890.
Repeated punctuations!!!!! should be removed, along with **random symbols** like @@,##.
stopwords like "the", "is", and "a" appear often.
HTML tags might be present: <div>This is inside a div</div>
And sometimes, contractions won't expand: "can't", "won't", "shouldn't".
Random numeric values: 123456, 98765, 2024.
"""

# 1Ô∏è‚É£ Remove HTML tags
text = re.sub(r'<.*?>', '', text)

# 2Ô∏è‚É£ Remove emails, phone numbers, and prices
text = re.sub(r'\S+@\S+', '', text)                     # emails
text = re.sub(r'\(?\b\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}\b', '', text)  # phone numbers
text = re.sub(r'\$?\d+(?:[\.,]\d+)?(?:\s?USD)?', '', text)  # prices

# 3Ô∏è‚É£ Remove punctuation, symbols, emojis, and numbers
text = re.sub(r'[^\w\s]', ' ', text)                    # keep letters/numbers
text = re.sub(r'[\U00010000-\U0010ffff]', '', text)     # remove emojis
text = re.sub(r'\d+', '', text)                         # remove numbers

# 4Ô∏è‚É£ Normalize spaces
text = re.sub(r'\s+', ' ', text).strip()

# 5Ô∏è‚É£ Convert to lowercase
clean_text = text.lower()

# 6Ô∏è‚É£ Remove basic stopwords manually
stopwords = {'the', 'is', 'a', 'and', 'of', 'to', 'in', 'it', 'this', 'that', 'on', 'for', 'be'}
words = [word for word in clean_text.split() if word not in stopwords]
clean_text = ' '.join(words)

# 7Ô∏è‚É£ Print or save result
print(clean_text)

with open("cleaned_text.txt", "w") as f:
    f.write(clean_text)
```
